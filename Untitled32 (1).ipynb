```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
```

```python
Missing_Values = ["?"," ","_","NaN","nan"]

home = pd.read_csv('Nashville_housing_data_2013_2016.csv', na_values= Missing_Values)
home.head()
```

```python
home.columns
```

```python
home= home[['Land Use', 'Sale Price', 'Multiple Parcels Involved in Sale', 'Acreage', 'Neighborhood',
       'Land Value', 'Building Value', 'Total Value', 'Finished Area', 'Year Built', 
        'Bedrooms', 'Full Bath', 'Half Bath']]
home
```

```python
home.isnull().sum()
```

```python
home = home.dropna(thresh=len(home.columns)-5)
home= home.dropna(subset=['Half Bath'])
home.isnull().sum()
```

```python
sns.boxplot(data=home, x="Finished Area").set(title='Finished Area Boxplot Distribution')
print("Median: ",home["Finished Area"].median())
print("Mean: ",home["Finished Area"].mean())
```

```python
home['Bedrooms'].value_counts().plot(kind='barh')
```

```python
home['Year Built'].value_counts().plot(kind='barh')
```

```python
plt.hist(home['Sale Price'])
```

```python
home["Finished Area"] = home["Finished Area"].fillna(home["Finished Area"].mean())
home.isnull().sum()
```

```python
sns.boxplot(data=home, x="Year Built").set(title='Boxplot Distribution of Built Year')
print("Median: ",home["Year Built"].median())
print("Mean: ",home["Year Built"].mean())
```

```python
home["Year Built"] = home["Year Built"].fillna(home["Year Built"].median())
home.isnull().sum()
```

```python
sns.boxplot(data=home, x="Bedrooms").set(title='Boxplot Distribution of Bedrooms')
print("Median: ",home["Bedrooms"].median())
print("Mean: ",home["Bedrooms"].mean())
```

```python
home["Bedrooms"] = home["Bedrooms"].fillna(home["Bedrooms"].median())
home.isnull().sum()
```

```python
sns.jointplot(x='Bedrooms',y='Sale Price',data=home,kind='scatter',size=4)
```

```python
print(len(home['Land Use'].unique()))
home['Land Use'].unique()
```

```python
home = home.replace(['SINGLE FAMILY', 'VACANT RES LAND', 'DUPLEX', 'ZERO LOT LINE',
       'TRIPLEX', 'RESIDENTIAL COMBO/MISC', 'QUADPLEX','DORMITORY/BOARDING HOUSE', 
        'SPLIT CLASS', 'MOBILE HOME', 'VACANT RESIDENTIAL LAND',
         'APARTMENT: LOW RISE (BUILT SINCE 1960)'],'Residential')

home = home.replace(['CHURCH', 'STRIP SHOPPING CENTER','PARSONAGE', 'OFFICE BLDG (ONE OR TWO STORIES)',
                'CONVENIENCE MARKET WITHOUT GAS', 'CLUB/UNION HALL/LODGE', 'LIGHT MANUFACTURING',
                'ONE STORY GENERAL RETAIL STORE', 'DAY CARE CENTER', 'TERMINAL/DISTRIBUTION WAREHOUSE', 
                 'NON-PROFIT CHARITABLE SERVICE',],'Commercial')

home = home.replace(['FOREST', 'GREENBELT/RES\r\nGRRENBELT/RES','GREENBELT', ],'Green_Zone')
home['Land Use'].unique()
```

```python
home = pd.get_dummies(home)
home.head()
```

```python
cor = home.corr()
plt.figure(figsize=(10 , 12))
heatmap = sns.heatmap(cor, vmin=-1, vmax=1, annot=True, cmap='YlGnBu')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
```

```python
X = home.drop(['Sale Price'], axis=1)
X
```

```python
Y = home["Sale Price"]
Y
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test,Y_Train, Y_Test = train_test_split(X, Y, test_size=0.2)
print("Train:", X_Train.shape, Y_Train.shape)
print("Test:", X_Test.shape, Y_Test.shape)
```

```python
from sklearn import linear_model
lm1 = linear_model.LinearRegression().fit(X_Train, Y_Train)
Y_Pred = lm1.predict(X_Test)
```

```python
print('Intercept: ', lm1.intercept_)
print('Coefficients: ', lm1.coef_)
```

```python
X_Train
```

```python
from sklearn.metrics import mean_squared_error
print('Mean Squared Error(MSE):%.2f' % mean_squared_error(Y_Test,Y_Pred))

from sklearn.metrics import r2_score
print('r_sqrd value:%.3f' % r2_score(Y_Test,Y_Pred))
```

```python
abs_co = abs(lm1.coef_)
abs_co
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import explained_variance_score
from sklearn.metrics import confusion_matrix
```

```python
from sklearn import metrics
from sklearn import tree

DT1 = tree.DecisionTreeClassifier().fit(X_Train, Y_Train)
Y_Pred = DT1.predict(X_Test)
print("Accuracy:",metrics.accuracy_score(Y_Test, Y_Pred))
```

```python
import sklearn
from sklearn.metrics import classification_report
#sklearn.metrics.classification_report
report = classification_report(Y_Pred, Y_Test)
print(report)
```

```python
#dt
tr_regressor = DecisionTreeRegressor(random_state=0)
tr_regressor.fit(X_Train,Y_Train)
tr_regressor.score(X_Test,Y_Test)
pred_tr = tr_regressor.predict(X_Test)
decision_score=tr_regressor.score(X_Test,Y_Test)
expl_tr = explained_variance_score(pred_tr,Y_Test)
```

```python
expl_tr
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees


print('Training Features Shape:', X_Train.shape)
print('Training Labels Shape:', X_Test.shape)
print('Testing Features Shape:', Y_Train.shape)
print('Testing Labels Shape:', Y_Test.shape)
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(X_Train, Y_Train);
```

```python
mlr = LinearRegression()
mlr.fit(X_Train,Y_Train)
mlr_score = mlr.score(X_Test,Y_Train)
pred_mlr = mlr.predict(X_Test)
expl_mlr = explained_variance_score(pred_mlr,Y_Test)
```

```python
rf_regressor = RandomForestRegressor(n_estimators=28,random_state=0)
rf_regressor.fit(X_Train,Y_Train)
rf_regressor.score(X_Test,Y_Test)
rf_pred =rf_regressor.predict(X_Test)
rf_score=rf_regressor.score(X_Test,Y_Test)
expl_rf = explained_variance_score(rf_pred,Y_Test)
```

```python
rf_regressor
rf_score
```

```python
from sklearn.tree import DecisionTreeRegressor
print("Multiple Linear Regression Model Score is ",round(mlr.score(X_Test,Y_Test)*100))
print("Decision tree  Regression Model Score is ",round(tr_regressor.score(X_Test,Y_Test)*100))
print("Random Forest Regression Model Score is ",round(rf_regressor.score(X_Test,Y_Test)*100))

#Let's have a tabular pandas data frame, for a clear comparison

models_score =pd.DataFrame({'Model':['Multiple Linear Regression','Decision Tree','Random forest Regression'],
                            'Score':[mlr_score,decision_score,rf_score],
                            'Explained Variance Score':[expl_mlr,expl_tr,expl_rf]
                           })
models_score.sort_values(by='Score',ascending=False)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.ensemble import GradientBoostingClassifier

grid = {

    'learning_rate':[0.01,0.05,0.1],

    'n_estimators':np.arange(100,500,100),

}


gb = GradientBoostingClassifier()

gb_cv = GridSearchCV(gb, grid, cv = 4)

gb_cv.fit(X_Train,Y_Train)

print("Best Parameters:",gb_cv.best_params_)

print("Train Score:",gb_cv.best_score_)

print("Test Score:",gb_cv.score(X_Test,X_Test))
```
